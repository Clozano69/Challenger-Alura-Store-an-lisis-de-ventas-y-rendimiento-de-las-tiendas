### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print("Ingreso Tienda:", ingreso_tienda)
print("Ingreso Tienda 2:", ingreso_tienda2)
print("Ingreso Tienda 3:", ingreso_tienda3)
print("Ingreso Tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
ventas_categoria_tienda = tienda["Categoría del Producto"].value_counts()
ventas_categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()
```

```python
print("Ventas por categoría - Tienda")
print(ventas_categoria_tienda)

print("\nVentas por categoría - Tienda 2")
print(ventas_categoria_tienda2)

print("\nVentas por categoría - Tienda 3")
print(ventas_categoria_tienda3)

print("\nVentas por categoría - Tienda 4")
print(ventas_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()
```

```python
print("Calificación promedio tienda:", round(promedio_tienda, 2))
print("Calificación promedio tienda 2:", round(promedio_tienda2, 2))
print("Calificación promedio tienda 3:", round(promedio_tienda3, 2))
print("Calificación promedio tienda 4:", round(promedio_tienda4, 2))
```

# 4. Productos más y menos vendidos

```python
ventas_producto_tienda = tienda["Producto"].value_counts()
ventas_producto_tienda2 = tienda2["Producto"].value_counts()
ventas_producto_tienda3 = tienda3["Producto"].value_counts()
ventas_producto_tienda4 = tienda4["Producto"].value_counts()
```

```python
print("Producto más vendido tienda:", ventas_producto_tienda.idxmax())
print("Producto menos vendido tienda:", ventas_producto_tienda.idxmin())

print("\nProducto más vendido tienda 2:", ventas_producto_tienda2.idxmax())
print("Producto menos vendido tienda 2:", ventas_producto_tienda2.idxmin())

print("\nProducto más vendido tienda 3:", ventas_producto_tienda3.idxmax())
print("Producto menos vendido tienda 3:", ventas_producto_tienda3.idxmin())

print("\nProducto más vendido tienda 4:", ventas_producto_tienda4.idxmax())
print("Producto menos vendido tienda 4:", ventas_producto_tienda4.idxmin())
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda = tienda["Costo de envío"].mean()
envio_promedio_tienda2 = tienda2["Costo de envío"].mean()
envio_promedio_tienda3 = tienda3["Costo de envío"].mean()
envio_promedio_tienda4 = tienda4["Costo de envío"].mean()
```

```python
print("costo de envío promedio tienda:", round(envio_promedio_tienda, 2))
print("costo de envío promedio tienda2:", round(envio_promedio_tienda2, 2))
print("costo de envío promedio tienda3:", round(envio_promedio_tienda3, 2))
print("costo de envío promedio tienda4:", round(envio_promedio_tienda4, 2))
```

# 6. Gráficas

```python
import matplotlib.pyplot as plt
```

Gráfica de ingresos

```python
ingresos = [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

tiendas = ["Tienda ", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, ingresos)

plt.title("Ingresos por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

plt.show()
```

Gráfica por calificasión promedio

```python
calificaciones = [
    promedio_tienda,
    promedio_tienda2,
    promedio_tienda3,
    promedio_tienda4
]

tiendas = ["Tienda", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.plot(tiendas, calificaciones, marker="o")

plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación")

plt.show()
```

Gráfica pór categoria de ventas

```python
ventas_categoria_tienda.plot(kind="pie")

plt.title("Ventas por categoría - Tienda")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de ventas")

plt.show()
```

Gráfica por envío promedio

```python
envios = [
    envio_promedio_tienda,
    envio_promedio_tienda2,
    envio_promedio_tienda3,
    envio_promedio_tienda4
]

plt.bar(tiendas, envios)

plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo de envío")

plt.show()
```

# 7. Informe final de análisis por tienda


INTRODUCCIÓN

El objetivo de este análisis fue evaluar el desempeño de cuatro tiendas con el fin de determinar cuál debería vender el Sr. Juan. Para ello se analizaron diferentes variables como los ingresos totales, las categorías de productos más vendidas, las calificaciones promedio de los clientes, los productos más y menos vendidos y el costo promedio de envío.

Además, se realizaron visualizaciones utilizando la librería Matplotlib para facilitar la interpretación de los resultados y comparar el desempeño entre las diferentes tiendas.

DESARROLLO DEL ANÁLISIS

Ingresos totales:

En primer lugar, se analizaron los ingresos totales de cada tienda, lo que permitió identificar cuáles generan mayores ventas. Este indicador es importante porque muestra el desempeño financiero general de cada tienda a la hora de tomar una decisión.

Categoría de productos:

Posteriormente, se analizaron las categorías de productos más y menos vendidas, lo que permitió identificar cuáles son los tipos de productos con mayor demanda entre los clientes. Este análisis ayuda a comprender mejor las preferencias del mercado y enfocar el esfuerzo financiero y estratégico en esos productos y obtener mayor rotación de los mismos

Calificación promedio:.

También se calcularon las calificaciones promedio de los clientes, lo que permite medir el nivel de satisfacción de los compradores con los productos y el servicio ofrecido por cada tienda, esto es un análisis importante ya que al saber cual tienda presenta mayor interés entre los usuarios permitirá mayor visita y por tanto mayores ventas de esa tienda.

Productos más vendidos:

Adicionalmente, se identificaron los productos más y menos vendidos en cada tienda, lo que permite entender cuáles tienen mayor aceptación entre los consumidores, en el entorno empresarial y de los negocios tener una bodega donde el producto no rote es una perdida de dinero, este análisis nos ayuda a direccionar la inversión a los proveedores y producción más necesaria para obtener un retorno de inversión más rápido, a su vez haya un gran flujo de caja.

Costo promedio de envío:

Por último, se analizó el costo promedio de envío en cada tienda. Aunque este costo lo paga el cliente, sigue siendo un factor importante ya que puede influir en la decisión de compra.

Para facilitar la interpretación de estos resultados, se generaron diferentes gráficos que permiten visualizar mejor las diferencias entre las tiendas.

CONCLUSÍON

A partir de estos resultados, se observa que la Tienda 4 presenta un desempeño inferior en comparación con las demás tiendas. Sus ingresos son menores, las calificaciones de los clientes son más bajas y algunos de sus productos tienen menor volumen de ventas en comparación con las otras tiendas analizadas.

Además, aunque el costo de envío puede variar entre tiendas, este factor no compensa las diferencias observadas en ingresos y satisfacción de los clientes.

Por esta razón, se recomienda que el Sr. Juan considere vender la Tienda 4, ya que presenta menores indicadores de desempeño en relación con las otras tiendas. Esta decisión permitiría enfocar los recursos en las tiendas que muestran mejores resultados en términos de ventas para mayor ganancia y satisfacción del cliente.
